```python
import numpy as np
from sklearn.datasets import make_hastie_10_2
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_friedman1
```

```python
# Exemplo de Regressão

#make_friedman1: Gera um conjunto de dados sintético para regressão com 10 variáveis explicativas. É um benchmark clássico para testar modelos de regressão.

X, y = make_friedman1(n_samples=1200, random_state=0, noise=1.0)
X_train, X_test = X[:200], X[200:]
y_train, y_test = y[:200], y[200:]
est = GradientBoostingRegressor(
    n_estimators=100, learning_rate=0.1, max_depth=1, random_state=0,
    loss='squared_error' #Usa mean_squared_error para medir o desempenho do modelo no conjunto de teste.

).fit(X_train, y_train)
mean_squared_error(y_test, est.predict(X_test))

print(mean_squared_error(y_test, est.predict(X_test)))
#Cálculo do MSE: ~5
# Um valor de ~5 indica que o modelo está capturando bem a estrutura dos dados.
```

```python
# Exemplo de Classificação

#make_hastie_10_2: Gera um conjunto de dados sintético para classificação binária com 10 variáveis.

X, y = make_hastie_10_2(random_state=0)
X_train, X_test = X[:2000], X[2000:]
y_train, y_test = y[:2000], y[2000:]

clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
    max_depth=1, random_state=0).fit(X_train, y_train)
clf.score(X_test, y_test)

print(clf.score(X_test, y_test))
#Acurácia do modelo: Quanto mais próximo de 1.0, melhor.
```
