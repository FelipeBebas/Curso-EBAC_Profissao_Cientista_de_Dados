# Cálculo - Tarefa 01

1. Marque quais desses métodos/algoritmos muito populares em ciência de dados são baseados no uso de derivada:

    1. Método Mínimos Quadrados
    2. Gradiente descendente
    3. Newton Raphson
    4. CART (Árvore de decisão)

**Os métodos 1,2,3 são baseados no uso de derivada. Já o método 4 (CART) é baseado em divisão de dados, como entropia ou Gini.**

2. Dada uma base de dados com uma variável resposta $y$ e um conjunto de variáveis explicativas. Considere uma estrutura de um modelo de regressão. Explique com suas palavras por que não é possível obter parâmetros que forneçam um erro quadrático médio (EQM) menor que o obtido com estimadores de mínimos quadrados.

**O método dos mínimos quadrados fornece os estimadores ótimos dos coeficientes de regressão, no sentido de minimizar a soma dos quadrados dos resíduos.
Como o EQM é diretamente derivado dessa soma, nenhum outro conjunto de parâmetros pode gerar um EQM menor — a solução já é a melhor possível dentro do modelo linear.**
