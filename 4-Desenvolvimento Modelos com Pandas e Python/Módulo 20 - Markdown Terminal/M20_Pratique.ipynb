## Exercício 1: Fórmula de Bhaskara


---


<h2 style="color: rgb(46, 85, 116);">Fórmula de Bhaskara</h2>

A solução para uma função de segundo grau depende de suas raízes (valores de x).  
Os coeficientes precisam ser números reais e o termo quadrático (*a*) diferente de zero.  
Sendo assim, temos a seguinte fórmula de Bhaskara:

<br>

$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$
<p align="center"><i>fórmula de Bhaskara</i></p>

<br>

**Discriminante**

A figura dentro da raiz na fórmula de Bhaskara é chamada de **discriminante**.  
Seu símbolo é a letra grega delta (\( \Delta \)), e sua fórmula correta é:

<br>

$$
\Delta = b^2 - 4ac
$$

- Se \( \Delta > 0 \), a equação terá **duas raízes reais e distintas**.  
- Se \( \Delta = 0 \), a equação terá **uma raiz real (duas iguais)**.  
- Se \( \Delta < 0 \), a equação **não terá raízes reais**.
